```python
#importing libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('bird.jfif', 0)  #read image in grayscale

#PLOT for IMAGES
fig = plt.figure(figsize=(10,5))
#plot original image
fig.add_subplot(1, 2, 1)
plt.imshow(img, cmap = 'gray')
plt.axis('off')
plt.title("Original Image")

#plot equalised image
img_equ = cv2.equalizeHist(img)
fig.add_subplot(1, 2, 2)
plt.imshow(img_equ, cmap = 'gray')
plt.axis('off')
plt.title("Equalized Image")

#PLOT for IMAGE HISTOGRAMS
fig = plt.figure(figsize=(10,5))

#histogram plot for original image
fig.add_subplot(1, 2, 1)
plt.hist(img.flatten(), bins=256)
plt.xlabel("Pixel value")
plt.ylabel("Pixels frequency") 
plt.title("Histogram - Original Image")

#histogram plot for equalized image
fig.add_subplot(1, 2, 2)
plt.hist(img_equ.flatten(), bins=256)
plt.xlabel("Pixel value")
plt.ylabel("Pixels frequency") 
plt.title("Histogram - Equalized Image")

plt.show()
```

```python
from skimage import data, exposure #loading class for predefined histogram macth

img = cv2.imread('bird.jfif', 0)
reference = cv2.imread('reference.jpg', 0)

dim = (img.shape[1], img.shape[0])
# adjust dimension (of match image to original)
reference_resized = cv2.resize(reference, dim)

matched = exposure.match_histograms(img, reference_resized) #image matching

fig = plt.figure(figsize=(12,4))

# plot original image
fig.add_subplot(1, 3, 1)
plt.imshow(img, cmap = 'gray')
plt.axis('off')
plt.title("Image")

# plot reference image
fig.add_subplot(1, 3, 2)
plt.imshow(reference_resized, cmap = 'gray')
plt.axis('off')
plt.title("Reference Image")

# plot matched image (after histogram matching of input image to reference)
fig.add_subplot(1, 3, 3)
plt.imshow(matched, cmap = 'gray')
plt.axis('off')
plt.title("Matched Image")

plt.show()
```
