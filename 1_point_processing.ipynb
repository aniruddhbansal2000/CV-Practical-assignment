```python
#importing libraries
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('bird.jfif', 1)[:, :, ::-1]   # import colored image
print("Image shape is {}".format(img.shape))
l = img.max()  #get max pixel value (deduct from it)
img_neg = l - img  #invert value

fig = plt.figure(figsize=(15, 45))  #size for plot done
# image 1 (original)
fig.add_subplot(1, 3, 1)
plt.imshow(img)
plt.axis('off')
plt.title("Original Image")
  
# Adds a subplot for negative image
fig.add_subplot(1, 3, 2)
plt.imshow(img_neg)
plt.axis('off')
plt.title("Negative Image")

img = cv2.imread('bird.jfif', 0)  #read black-white image
m, n = img.shape

img_thresh = np.zeros((m,n), dtype = int)# threshold
T = 120  #cutoff
# generating threshold image
for i in range(m):
    for j in range(n):
        if(img[i,j] < T):
            img_thresh[i,j] = 0
        else:
            img_thresh[i,j] = 255
            
# Adds a subplot for threshold image
fig.add_subplot(1, 3, 3)
plt.imshow(img_thresh)
plt.axis('off')
plt.title("Threshold Image")

plt.show()
```
