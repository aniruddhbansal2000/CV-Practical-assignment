```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('bird.jfif')[:, :, ::-1]

fig = plt.figure(figsize=(5,5))
#plot original image
fig.add_subplot(1, 1, 1)
plt.imshow(img)
plt.axis('off')
plt.title("Image")
plt.show()

''' APPLYING DIFFERENT FILTERS ON THE IMAGE'''
kernel = np.ones((5,5), np.float32)/25

#MEAN FILTERING
'''employing predefined convolution function of cv2 and hence implementing mean filter'''
img_mean = cv2.filter2D(img, -1, kernel)

#add plot size for images obtained from different filters
fig = plt.figure(figsize=(10,7))

fig.add_subplot(2, 2, 1)
plt.imshow(img_mean)
plt.axis('off')
plt.title("Mean")


#MEDIAN FILTERING
img_median = cv2.medianBlur(img, 5)  #predefined via cv2
fig.add_subplot(2, 2, 2)
plt.imshow(img_median)
plt.axis('off')
plt.title("Median")

#LOW PASS FILTERING
kernel = (np.ones((5, 5))*4)/ 100
img_LPF = cv2.filter2D(img, -1, kernel)
fig.add_subplot(2, 2, 3)
plt.imshow(img_LPF)
plt.axis('off')
plt.title("Low-pass")

kernel = np.array([[0.0, -4.0, 0.0], [-4.0, 16.0, -4.0], [0.0, -4.0, 0.0]])
#HIGH PASS FILTERING
kernel = kernel/(np.sum(kernel) if np.sum(kernel) != 0 else 1)
img_HPF = cv2.filter2D(img, -1, kernel)
fig.add_subplot(2, 2, 4)
plt.imshow(img_HPF)
plt.axis('off')
plt.title("High-pass")
plt.show()
```
