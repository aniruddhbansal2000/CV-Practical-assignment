```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img1 = cv2.imread('img1.png')[:, :, ::-1]
img2 = cv2.imread('img2.png')[:, :, ::-1]

fig = plt.figure(figsize=(15,8))

fig.add_subplot(3, 3, 1)
plt.imshow(img1)
plt.axis('off')
plt.title("Image 1")

fig.add_subplot(3, 3, 2)
plt.imshow(img2)
plt.axis('off')
plt.title("Image 2")

# applying logical operators on image (ie their matrix representations)
img_and = cv2.bitwise_and(img1, img2, mask = None)
img_or = cv2.bitwise_or(img1, img2, mask = None)
img_xor = cv2.bitwise_xor(img1, img2, mask = None)
img1_not= cv2.bitwise_not(img1, mask = None)
img2_not= cv2.bitwise_not(img2, mask = None)

fig.add_subplot(3, 3, 4)

plt.imshow(img_and)  # plot AND image
plt.axis('off')
plt.title("AND")

fig.add_subplot(3, 3, 5)
plt.imshow(img_or)  # plot OR image
plt.axis('off')
plt.title("OR")

fig.add_subplot(3, 3, 6)
plt.imshow(img_xor)  # plot XOR image
plt.axis('off')
plt.title("XOR")

fig.add_subplot(3, 3, 7)
plt.imshow(img1_not)  # plot IMG1-NOT image
plt.axis('off')
plt.title("NOT of Image-1")

fig.add_subplot(3, 3, 8)
plt.imshow(img2_not) # plot IMG2-NOT image
plt.axis('off')
plt.title("NOT of Image-2")

plt.show()
```

```python
from PIL import Image

# text Watermark
from PIL import ImageFont
from PIL import ImageDraw

img = Image.open("bird.jfif")
watermark_image = img.copy()
  
draw = ImageDraw.Draw(watermark_image)  #the base image
font = ImageFont.truetype("arial.ttf", 50)  #font to add
  
fig = plt.figure(figsize=(10,5))

# add watermark (black text)
draw.text((0, 0), "bird", 
          (0, 0, 0), font=font)
plt.subplot(1, 2, 1)
plt.axis('off')
plt.title("black text")
plt.imshow(watermark_image)
  
# add watermark (white text)
draw.text((0, 0), "bird", 
          (255, 255, 255), font=font)
plt.subplot(1, 2, 2)
plt.axis('off')
plt.title("white text")
plt.imshow(watermark_image)
plt.show()
```
